```python
c = df_rape.groupby(['Area_Name'])['Victims_of_Rape_Total'].sum().reset_index()
c
#looking at the rape cases as per state
```

```python
# Plotting with total number of Rapes Statewise
plt.figure(figsize=(50,15))
plt.xlabel("Area_Name", fontsize=40)
plt.ylabel("Victims_of_Rape_Total", fontsize=40)
plt.xticks(rotation=90,fontsize=30)
plt.yticks(rotation=0,fontsize=20)
sns.barplot(x="Area_Name",y="Victims_of_Rape_Total",data=df_rape)
#highest in uttar pradesh,lowest in daman and diu
```

```python
#plotting the graph for old peoplewho were victims
fig = plt.figure(figsize = (15,15))
plt.barh(area,old)
plt.title('State Rape Victims who were above 50 age')
plt.xlabel('Rape_Cases_Reported')
plt.ylabel('State')
plt.show()
                      
```

```python
state = df_rape.groupby('Area_Name').sum()
area = df_rape['Area_Name']
old = df_rape['Victims_Above_50_Yrs']
teen1 = df_rape['Victims_Between_10-14_Yrs']
teen2 = df_rape['Victims_Between_14-18_Yrs']
teen3 = df_rape['Victims_Upto_10_Yrs']
teen = teen1+teen2+teen3
```

```python
fig = plt.figure(figsize = (15,15))
plt.barh(area,teen)
plt.title('State Rape Victims who were below 18 years of age')
plt.xlabel('Rape_Cases_Reported')
plt.ylabel('State')
plt.show()
#plotting graphs for teen who were the victims          
```

```python
fig = plt.figure(figsize = (15,15))
plt.barh(area,adult)
plt.title('State Rape Victims who were above 18 years and below 50 years')
plt.xlabel('Rape_Cases_Reported')
plt.ylabel('State')
plt.show()
```

```python
df_rape['Unreported'] = df_rape['Victims_of_Rape_Total'] - df_rape['Rape_Cases_Reported']
df_rape['Unreported'].max()
#highest number of unreported rape cases where 12
```

```python
rape_victims_by_state = df_rape.groupby('Area_Name').sum()
rape_victims_by_state.drop('Year', axis = 1, inplace = True)
print('Total Rape Victims = ' ,rape_victims_by_state['Rape_Cases_Reported'].sum())
rape_victims_by_state.sort_values(by = 'Rape_Cases_Reported', ascending = False).head()
#printing total number of rape cases
```
